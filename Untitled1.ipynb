```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import heapq
```

```python
data = np.load('ihb.npy')
data
```

```python
data[1,0,200:]
```

```python
# 1) берем 5 максимумов и 5 минимумов (10 элементов)
# 2) проходим 320 раз по первому кадру на оси 246, находя модуль разности, заносим это в temp, затем суммируем
# и получаем массив (320, 1)
# 3) прогоняем в цикле 1 и 2 пункты по оси 10, получаем массив (320, 10)
# 4) суммируем ось 10 и получаем массив (320, 1)
```

```python
data = data[:,:,:-46]
```

```python
data_temp = data
data_twins = {}
```

```python
for b in range(0,20):
    stat_data_10 = []

    if data_temp.shape[0] > 1:
        for n in range(0,10):
            stat_data_246 = []
            temp = []
            
            largest_with_indices = heapq.nlargest(10, enumerate(data_temp[0,n,:]), key=lambda x: x[1])
            smallest_with_indices = heapq.nsmallest(10, enumerate(data_temp[0,n,:]), key=lambda x: x[1])
            # Извлекаем только индексы
            indices = [index for index, value in largest_with_indices]
            temp.extend(indices)
            #print(temp)
            indices = [index for index, value in smallest_with_indices]
            temp.extend(indices)
            
            for i in range(0,data_temp.shape[0]):
                a = 0
                for q in temp:
                    a = a + abs(data_temp[0,n,q] - data_temp[i,n,q])
                stat_data_246.append(a)
            stat_data_10.append(stat_data_246)
        #print(stat_data_246[1])
        stat_data_10 = np.array(stat_data_10)
        #print(stat_data_10.shape)
        stat_data_10 = stat_data_10.sum(axis=0)
        stat_data_10 = stat_data_10.tolist()
        # на этом этаме в stat_data_10 массим из 320 значений, где по оси 246 и оси 10 находятся коэфициенты сравнения графиков оси 320 с первым графиком (0).
        smallest_temp = heapq.nsmallest(16,stat_data_10)[1:]
        temp_16 = []
        temp_16_idx = []
        # заполняем массив "близнецов"
        for v in range(0,15):
            temp_16.append(np.where(data[:,0,0] == data_temp[stat_data_10.index(smallest_temp[v]),0,0])[0][0])
                  
        data_twins.update([(b, [np.where(data[:,0,0] == data_temp[0,0,0])[0][0], *temp_16  ])])          
        # удаляем предыдущих "близнецов"
        data_temp = np.delete(data_temp, 0, axis=0)
        #print(temp_16) 
        
        for z in temp_16:
            temp_16_idx.append(np.where(data_temp[:,0,0] == data[z,0,0])[0][0])
 
        data_temp = np.delete(data_temp, temp_16_idx, axis=0)
```

```python
df_twins = pd.DataFrame(data_twins)
df_twins
```
