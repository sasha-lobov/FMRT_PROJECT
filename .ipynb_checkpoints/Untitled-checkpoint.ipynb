```python
filtered_data = []
temp = []
for q in range(0,10):
    for i in range(0,320):
        
        df = pd.DataFrame(data[i,q,:], columns=['value'])
        
        #  IQR
        Q1 = df['value'].quantile(0.25)
        Q3 = df['value'].quantile(0.75)
        IQR = Q3 - Q1
        
        #    
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        #  
        temp.append(list(df[(df['value'] >= lower_bound) & (df['value'] <= upper_bound)]))
    filtered_data.append(list(temp))
    temp = []
    
```
